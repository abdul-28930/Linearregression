# Using car dataset and implementing Linear regression 

### Importing the required libraries


```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.utils import shuffle
from sklearn import linear_model
import matplotlib.pyplot as plt
from matplotlib import style
import pickle
```

### Importing the required dataset

```python
data = pd.read_csv("cars.csv")
```

### Selecting needed attributes only

```python
data = data[["price", "horsepower", "enginesize", "stroke", "curbweight"]]
```

### Assume a new variable and create 2 arrays

```python
predict = "price"

x = np.array(data.drop([predict], axis=1))
y = np.array(data[predict])
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, train_size=0.3)
```

### Assume best as 0 so we can run a for loop on it and get the best accuracy

```python
best = 0
```

### Run the for loop, and get the best accuracy and store it in a pickle file

```python
for _ in range(10):
    x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, train_size=0.3)

    linear = linear_model.LinearRegression()

    linear.fit(x_train,y_train)
    accuracy = linear.score(x_test,y_test)
    print("Accuracy : \n", accuracy)
    
    if(accuracy>best):
        best=accuracy
        with open ("car.pickle", "wb") as f:
            pickle.dump(linear, f)
```

### Now, open the pickle file by loading it

```python
pick_in = open("car.pickle", "rb")
linear = pickle.load(pick_in)
```

### Now, let's find the coefficient and the intercept for it

```python
print("Coeffieicient = \n", linear.coef_)
print("intercept = \n", linear.intercept_)
```

### Now, let's check our predictions

```python
predictions = linear.predict(x_test)

for x in range(len(predictions)):
    print(predictions[x], x_test[x], y_test[x])
```
